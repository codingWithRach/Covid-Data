```python
# import packages
import numpy as np
import pandas as pd
import get_data as gd
import display_data as dd
```

```python
# import and tidy up data
cases_df = gd.get_cases_data()
cases_by_age_df = gd.get_cases_by_age_data()

# create data frame of cases restricted to the most recent date
cases_max_date_df = cases_df[cases_df.date == cases_df.date.max()]
```

```python
# display the rows with the highest rolling rate
dd.show_latest_cases(cases_max_date_df.sort_values('rollingrate', ascending=False), 15)
```

```python
# display bubble chart for rolling rate
dd.bubble_chart(cases_max_date_df, 'rollingrate', 'rate', 'Rolling Rate', 10)
```

```python
# display bubble chart for rolling sum
dd.bubble_chart(cases_max_date_df, 'rollingsum', 'sum', 'Rolling Sum', 10)
```

```python
# Restrict data to that corresponding to an area generated at random, and ensure it's in date order
areaname = dd.get_random_areaname(cases_by_age_df)
df_area = dd.get_df_area(cases_by_age_df, areaname).sort_values('date')
```

```python
# Display graph for rolling sum
col_names = ['rollsum_0_49', 'rollsum_50_69', 'rollsum_70_89', 'rollsum_90+']
labels = ['0 to 49', '50 to 69', '70 to 89', '90 Plus']
dd.plot_cases(df_area, col_names, labels, "Cases in " + areaname + " by age","Rolling Sum")
```

```python
# Display graph for rolling rate
col_names = ['rollrate_0_49', 'rollrate_50_69', 'rollrate_70_89', 'rollrate_90+']
labels = ['0 to 49', '50 to 69', '70 to 89', '90 Plus']
dd.plot_cases(df_area, col_names, labels, "Cases per 100,000 in " + areaname + " by age","Rolling Rate")
```
